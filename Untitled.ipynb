## Model building

```python
from sklearn.metrics import make_scorer
```

```python
def r_score1(y_pred,y):
    
    y_pred = np.log(y_pred)
    y = np.log(y)
    return 1 - ((np.sum((y_pred - y)**2)) / len(y))**1/2

# SCORING func for Cross Validation (RandomizedSearchCV)
score = make_scorer(score_func = r_score1, greater_is_better=True)
```

```python
LG_reg = LinearRegression()
DC_reg = DecisionTreeRegressor()
KNN_reg = KNeighborsRegressor()
SVR_reg = SVR()
RAN_reg = RandomForestRegressor()
```

```python
for reg in (LG_reg, DC_reg , KNN_reg, SVR_reg, RAN_reg):
    reg.fit(x_train , y_train)
    
    y_pred = reg.predict(x_test)
    
    print(reg, score(y_pred,y_test))
```
